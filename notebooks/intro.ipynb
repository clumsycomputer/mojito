```mojo
from loop import Loop, LoopLayer, LOOP_ONE, LOOP_ZERO, LOOP_PI
from python import Python
Plot = Python.import_module('matplotlib.pyplot')

var current_loop = Loop[layer_count=2](
    InlineArray[LoopLayer, 2](
        LoopLayer(
            relative_sub_loop_radius=0.75,
            relative_sub_loop_depth=0.75,
            sub_phase=LOOP_PI / 2,
            sub_orientation=0,
            loop_rotation=0,
        ),
        LoopLayer(
            relative_sub_loop_radius=0.675,
            relative_sub_loop_depth=0.675,
            sub_phase=LOOP_PI / 2,
            sub_orientation=0,
            loop_rotation=0,
        )
    )
)
var sample_count = 100
var samples_angle = Python.list()
var samples_cosine = Python.list()
var samples_sine = Python.list()
for sample_index in range(sample_count):
    var sample_angle = 2 * LOOP_PI / (sample_count - 1) * sample_index
    var sample_point = current_loop.pointAt(sample_angle)
    samples_angle.append(sample_angle)
    samples_cosine.append(sample_point.cosine())
    samples_sine.append(sample_point.sine())

Plot.plot(samples_angle, samples_sine)
Plot.title('Loop Sine')
Plot.xlabel('angle')
Plot.ylabel('sine')
Plot.grid(True)
Plot.show()

Plot.plot(samples_angle, samples_cosine)
Plot.title('Loop Cosine')
Plot.xlabel('angle')
Plot.ylabel('cosine')
Plot.grid(True)
Plot.show()

Plot.figure()
Plot.plot(samples_cosine, samples_sine)
Plot.title('Loop Shape')
Plot.axis('equal')
Plot.xlim(-1.5, 1.5)
Plot.ylim(-1.5, 1.5)
Plot.grid(True)
Plot.show()
```
