```mojo
from python import Python
from math import sin

Plot = Python.import_module('matplotlib.pyplot')
alias pi = 3.14
alias num_samples = 100
var samples_angle = Python.list()
var samples_sine = Python.list()

for sample_index in range(num_samples):
    var current_angle = 2 * pi / num_samples * sample_index
    var current_sine = sin(current_angle)
    samples_angle.append(current_angle)
    samples_sine.append(current_sine)

Plot.plot(samples_angle, samples_sine)
Plot.title('Sine Wave')
Plot.xlabel('x')
Plot.ylabel('sin(x)')
Plot.grid(True)
Plot.show()
```

```mojo
from python import Python
from math import cos, sin

Plot = Python.import_module('matplotlib.pyplot')
alias pi = 3.14
alias num_samples = 100
var samples_cos = Python.list()
var samples_sin = Python.list()

for sample_index in range(num_samples):
    var current_angle = 2 * pi / (num_samples - 1) * sample_index
    samples_cos.append(cos(current_angle))
    samples_sin.append(sin(current_angle))

Plot.figure()
Plot.plot(samples_cos, samples_sin)
Plot.axis('equal')
Plot.xlim(-1.1, 1.1)
Plot.ylim(-1.1, 1.1)
Plot.grid(True)
Plot.show()
```
